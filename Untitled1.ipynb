```python
import pandas as pd
```

```python
df = pd.read_csv('final_constituencies.csv')
```

```python
# Keep only these three columns
df_subset = df[['CandidateName', 'EnglishCandidateName', 'DistrictName', 'EnglishDistrictName', 'ConstName']]
```

```python
df_subset
```

```python
df = df_subset
# 1. Define the date variable for easy updates
search_date = "2025-09-05"

# 2. Create the PowerSearchQuery column
df['PowerSearchQuery'] = df.apply(lambda row: 
    f'("" {row["CandidateName"]} "" OR "" {row["EnglishCandidateName"]} "") '
    f'("" {row["DistrictName"]} "" OR "" {row["EnglishDistrictName"]} {row["ConstName"]} "") '
    f'since:{search_date}', axis=1)

# Display the first few rows to verify
df[['CandidateName', 'PowerSearchQuery']].head()
```

```python
df.to_csv('updated_candidates.csv', index=False, encoding='utf-8-sig')
```
