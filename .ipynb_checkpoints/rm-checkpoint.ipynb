# Candidate Data Cleanup
Remove redundant columns: `NepaliSearchQuery`, `EnglishSearchQuery`, `FileSaveName`

```python
import pandas as pd

df = pd.read_csv('candidate_data.csv', encoding='utf-8-sig')
print(f'Loaded {len(df)} rows, {len(df.columns)} columns')
df.head()
```

```python
# Drop redundant derived columns
redundant = ['NepaliSearchQuery', 'EnglishSearchQuery', 'FileSaveName']
df_clean = df.drop(columns=redundant)

print(f'Remaining columns: {list(df_clean.columns)}')
df_clean.head(10)
```

```python
# Quick stats
print(f'Total candidates : {len(df_clean)}')
print(f'Unique districts : {df_clean["EnglishDistrictName"].nunique()}')
print(f'Unique constituencies: {df_clean["ConstName"].nunique()}')
df_clean['EnglishDistrictName'].value_counts()
```

```python
# Save cleaned file
df_clean.to_csv('candidate_data_clean.csv', index=False, encoding='utf-8-sig')
print('Saved: candidate_data_clean.csv')
```
