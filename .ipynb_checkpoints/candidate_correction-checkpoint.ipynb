# Candidate Data — Name Correction + Search Query Optimization

This notebook:
1. Lets you correct English spelling of candidate/district names interactively
2. Rebuilds `PowerSearchQuery` with optimized X (Twitter) advanced search syntax

**Optimized query format:**
```
("NepaliName" OR "EnglishName") ("districtname" OR "constituency") since:YYYY-MM-DD
```
- The `OR` between names catches both scripts
- Adding district + "constituency" keyword narrows results to political content about that area
- `since:` limits to relevant period

````python
import pandas as pd

df = pd.read_csv('candidate_data_clean.csv', encoding='utf-8-sig')
print(f'Loaded {len(df)} rows')
df.head(8)
````

## Step 1 — Interactive Name Correction
Run the cell below. For each row:
- Press **Enter** to keep the current value
- Type a correction and press **Enter** to update
- Type **`q`** to stop and save progress

````python
df_work = df.copy()
changes = 0

print(f"{'─'*65}")
print(f"  Name Correction  ({len(df_work)} candidates)")
print(f"  Enter = keep  |  type new name = correct  |  q = stop")
print(f"{'─'*65}")

for i, row in df_work.iterrows():
    nepali   = row['CandidateName']
    eng_name = row['EnglishCandidateName']
    district = row['EnglishDistrictName']
    const    = row['ConstName']

    print(f"\n[{i+1}/{len(df_work)}] Nepali: {nepali}")
    
    # --- Candidate name ---
    ans = input(f"  EnglishCandidateName [{eng_name}]: ").strip()
    if ans == 'q':
        print("Stopping early.")
        break
    if ans:
        df_work.at[i, 'EnglishCandidateName'] = ans
        changes += 1
        print(f"  ✓ Updated to: {ans}")

    # --- District name ---
    ans2 = input(f"  EnglishDistrictName  [{district}]: ").strip()
    if ans2 == 'q':
        print("Stopping early.")
        break
    if ans2:
        df_work.at[i, 'EnglishDistrictName'] = ans2
        changes += 1
        print(f"  ✓ Updated to: {ans2}")

print(f"\n{'─'*65}")
print(f"  Done — {changes} field(s) corrected")
print(f"{'─'*65}")
````

## Step 2 — Rebuild PowerSearchQuery (Optimized)

New format uses proper X/Twitter advanced search operators:
```
("NepaliName" OR "EnglishName") ("district" OR "constituency") since:YYYY-MM-DD
```
- Exact-phrase quotes `""` prevent partial matches
- `OR` is uppercase (required by Twitter)
- District name + literal word `constituency` catches political tweets even when the candidate isn't named directly

````python
SINCE_DATE = '2025-09-05'   # ← change if needed

def build_query(row):
    nepali   = row['CandidateName'].strip()
    eng_name = row['EnglishCandidateName'].strip()
    district = row['EnglishDistrictName'].strip().lower()
    const    = str(row['ConstName']).strip()

    # Name part: both scripts
    name_part = f'("{nepali}" OR "{eng_name}")'
    
    # Location/context part: district name + constituency keyword
    location_part = f'("{district}" OR "constituency")'

    return f'{name_part} {location_part} since:{SINCE_DATE}'

df_work['PowerSearchQuery'] = df_work.apply(build_query, axis=1)

print("Sample queries:")
for q in df_work['PowerSearchQuery'].head(5):
    print(' ', q)
````
